```python
url_1= "https://mtp.travel/api/users/"
url_2 ="/location-posts"
```

```python
df_initial = pd.read_csv("raw_datawithstate.csv", index_col = 0)
df_initial
```

```python
ids_to_scrape = df_initial["id"].tolist()
ids_to_scrape
```

```python

counter=0
for j in range(1711, 8140, 59):
    for i in range(j, j+59):
        try:
            query_url = url_1+str(ids_to_scrape[i])+url_2
            post_response = requests.get(query_url)
            post_json = post_response.json()
            post_data =post_json["data"]
            # location_data = MTP_json["users"]["data"]["location"]
            # MTP_data
            post_df = pd.DataFrame(post_data)
            # MTP_df_final = MTP_df["first_name","last_name","birthday","location"]
            post_df_final = post_df_final.append(post_df.filter(["user_id","post","location"], axis=1))
            print("we're on user # " + str(i))
        except:
            counter = counter+1
            print("something weird happened with User ID " + str(ids_to_scrape[i]) + " for the " + str(i)+ " time, on the " + str(j)+" loop")
            time.sleep(20)
post_df_final
```
