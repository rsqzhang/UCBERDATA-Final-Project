```python
import pandas as pd
```

```python
europe_posts_df= pd.read_csv("cleaned_posts_only_europe.csv")
europe_posts_df
```

```python
good_words =["good","enjoy","enjoyed","love","loved","amazing","like","great","wonderful","nice"]
```

```python
europe_posts_df.post[0][8]
```

```python
europe_posts_df["post"]=[x.split(" ") for x in europe_posts_df["post"]]
europe_posts_df
```

```python
europe_posts_df["tf"]=[pd.Series(x).isin(good_words).sum() for x in europe_posts_df.post]
europe_posts_df
```

```python
europe_post_final_group=europe_posts_df.groupby(by="user_id")
europe_post_final_group
```

```python
europe_post_final= europe_post_final_group.max()
europe_post_final
```

```python
europe_post_final.to_csv("User_posts_only_Europe.csv")
```
