```python
import pandas as pd
```

```python
only_europe_posts = pd.read_csv("User_posts_only_Europe_re.csv")
only_europe_posts
```

```python
europe_locations_df = pd.read_csv("Europe_list_locations.csv")
europe_locations_df
```

```python
merged_posts_locations = only_europe_posts.merge(europe_locations_df,how="inner", left_on="location",right_on="location_id")
merged_posts_locations
```

```python
users_df = pd.read_csv("raw_datawithstate.csv")
users_df
```

```python
merged_posts_locations_user = merged_posts_locations.merge(users_df,how="inner",left_on ="user_id",right_on="id")
final_data = merged_posts_locations_user.filter(["full_name","birthday","gender","location_y","Abbr","location_name","country_name","region_name"])
final_data = final_data.rename(columns={"location_y":"state","Abbr":"state_abbr"})
```
