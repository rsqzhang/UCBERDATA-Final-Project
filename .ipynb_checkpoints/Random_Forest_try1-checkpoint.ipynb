```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
```

```python
rf = RandomForestClassifier(n_estimators=200)
```

```python
data = pd.read_csv('final_data_forML.csv')
data = data.filter(["Age","gender","state","country_name"])
x_data=data.loc[:,["Age","gender","state"]]
y_data = data.loc[:,"country_name"]
y_data
```

```python
for x in x_data.gender.unique():
    x_data[x]=[1 if y == x else 0 for y in x_data.gender]
x_data
```

```python
for x in x_data.state.unique():
    x_data[x]=[1 if y == x else 0 for y in x_data.state]
x_data = x_data.drop(columns =["gender","state"])
x_data
```

```python
y_data
```

```python
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(x_data, y_data)
rf.score(x_data, y_data)
```
